Funções proprias para a conversão de datas em formato GPS

```python
import datetime
def convertHora(segundos):
    data_inicial = datetime.datetime.strptime("06/01/1980 00:00:00", "%d/%m/%Y %H:%M:%S")
    segundos_iniciais = data_inicial.timestamp()
    segundos_finais = segundos_iniciais + segundos
    return datetime.datetime.fromtimestamp(segundos_finais)
def horaBr(horaUTC):
    diferenca = datetime.timedelta(hours=3)
    return horaUTC - diferenca
def gpsToDate(segundos):
  hora = horaBr(convertHora(segundos))
  return hora.strftime('%d/%m/%Y %H:%M:%S')
segundos = 1164556817 #int(input("Digite o número de segundos: "))
data_final = convertHora(segundos)
print(f"A data e hora em UTC      são {data_final.strftime('%d/%m/%Y %H:%M:%S')}")
brasilia = horaBr(data_final)
print(f"A data e hora em Brasília são {brasilia.strftime('%d/%m/%Y %H:%M:%S')}")
```

## Atividade 2: Acesso ao banco de dados (GWpy)

Receber informações com base no horario do GPS


```python
from gwosc.datasets import event_gps
from gwpy.timeseries import TimeSeries
gps = event_gps('GW190412')
print("hora do evento no formato de gps:", gps)
segmento_data = (int(gps)-5, int(gps)+5)
print("Selecionar uma hora inicial e final como segmento:",segmento_data)
ldata = TimeSeries.fetch_open_data('L1', *segmento_data, verbose=True)
#Nomes disponiveis
#'G1' - GEO600
#'H1' - LIGO-Hanford
#'L1' - LIGO-Livingston
#'V1' - (Advanced) Virgo
#'K1' - KAGRA
print("Dados disponiveis para esse intervalo de tempo no L1:",ldata)
```

Realizar o plot dos dados previamente coletados

```python
plot = ldata.plot()
```

Usando transformada de Fourier, realizar o tratamento da frequencia

```python
fft = ldata.fft()
print(fft)
```

```python
plot = fft.abs().plot(xscale="log", yscale="log")
plot.show(warn=False)
```

Esse grafico entretanto não está correto pois a transformada funciona usando o pressuposto de que nossos dados são periodicos, isso ocasiona essa sensação de descontinuidade no gráfico. Para corrigir isso, precisamos aplicar uma janela antes de realizar a transformada.

```python
from scipy.signal import get_window
window = get_window('hann', ldata.size)
lwin = ldata * window
fftamp = lwin.fft().abs()
plot = fftamp.plot(xscale="log", yscale="log")
plot.show(warn=False)
```

As densas flutuações no gráfico acima em frequências maiores que 10 Hz parecem completamente aleatórias. Este é de fato o ruído intrínseco na estimativa do conteúdo espectral do sinal de uma única FFT. Podemos reduzir essas flutuações calculando a média de muitas estimativas do sinal FFT. Mas, sendo os componentes da FFT quantidades complexas, o que realmente queremos calcular são os seus módulos quadrados. Essa quantidade é chamada de Densidade Espectral de Potência (PSD), e o que acabamos de descrever é o método de estimativa de Welch. Também é conveniente expressar o resultado não como uma potência, mas como uma amplitude, extraindo a raiz quadrada da expressão acima. A quantidade resultante leva o nome de Densidade Espectral de Amplitude (ASD) do sinal. 

```python
asd = ldata.asd(fftlength=2, method="median")
plot = asd.plot()
plot.show(warn=False)
```

Vamos personalizar esta figura ampliando a faixa de frequência de interesse, ou seja, acima do ombro do ruído sísmico em 10 Hz e abaixo do corte na frequência de Nyquist, e redimensionando o eixo y de acordo.

```python
ax = plot.gca()
ax.set(xlim=(10, 1400), ylim=(1e-24, 1e-20))
plot
```

O ASD é uma ferramenta padrão usada para estudar a sensibilidade no domínio da frequência de um detector de ondas gravitacionais. Para os dados do LIGO-Livingston que carregamos, podemos ver grandes picos em certas frequências, incluindo:  
~300Hz  
~500Hz  
~1000Hz  
A página de linhas espectrais de O3 no GWOSC descreve uma série dessas características espectrais para O3, com algumas delas sendo impostas a nós e outras introduzidas deliberadamente para ajudar no controle do interferômetro.
Carregar mais dados permite que mais FFTs sejam calculados durante a estimativa do ASD, o que significa que as variações aleatórias são calculadas e podemos ver mais detalhes:

```python
ldata2 = TimeSeries.fetch_open_data('L1', int(gps)-512, int(gps)+512, cache=True)
lasd2 = ldata2.asd(fftlength=4, method="median")
plot = lasd2.plot()
ax = plot.gca()
ax.set_xlim(10, 1400)
ax.set_ylim(1e-24, 1e-20)
plot.show(warn=False)
```

Agora podemos ver mais algumas características, incluindo um pequeno aumento em cerca de 60 Hz, a frequência da rede elétrica nos EUA.

Para comparação, podemos carregar os dados do LIGO-Hanford e Virgo e plotá-los também:

```python
# get Hanford data
hdata2 = TimeSeries.fetch_open_data('H1', int(gps)-512, int(gps)+512, cache=True)
hasd2 = hdata2.asd(fftlength=4, method="median")
```

```python
# get Virgo data
vdata2 = TimeSeries.fetch_open_data('V1', int(gps)-512, int(gps)+512, cache=True)
vasd2 = vdata2.asd(fftlength=4, method="median")
```

```python
# and plot using standard colours
ax.plot(hasd2, label='LIGO-Hanford', color='gwpy:ligo-hanford')
ax.plot(vasd2, label='Virgo', color='gwpy:virgo')

# update the Livingston line to use standard colour, and have a label
lline = ax.lines[0]
lline.set_color('gwpy:ligo-livingston')  # change colour of Livingston data
lline.set_label('LIGO-Livingston')

ax.set_ylabel(r'Strain noise [$1/\sqrt{\mathrm{Hz}}$]')
ax.legend()
plot
```

Exercicio 2: Make an ASD around the time of an O1 event, GW150914 for L1 detector.  Compare this with the ASDs around GW190412 for L1 detector.  Which data have lower noise - and so are more sensitive - around 100 Hz?

## Atividade 3: Q-transform e Spectrogram

Primeiro, vamos recarregar as informações sobre algumas das ondas

```python
from gwosc.datasets import event_gps
from gwpy.timeseries import TimeSeries

gps = event_gps('GW170817')
print("GW170817 GPS:", gps)

ldata = TimeSeries.fetch_open_data('L1', int(gps)-512, int(gps)+512, cache=True)
print("GW170817 data")
print(ldata)
```

Agora, vamos gerar o spectrogram usando FFT, já com a janela aplicada

```python
specgram = ldata.spectrogram2(fftlength=4, overlap=2, window='hann') ** (1/2.)
plot = specgram.plot()
```

Entretanto nenhum ponto está visivel, isso acontece porque a escala padrão de cores não é adequada para executar esses valores, por causa disso faremos um ajuste dela e para isso bata olhar para os valores minimos e maximos:

```python
print("Valor minimo:",specgram.min(), "Valor maximo:",specgram.max())
```

Ajustando esse grafico:

```python
ax = plot.gca()
ax.set_yscale('log')
ax.set_ylim(10, 1400)
ax.colorbar(
    clim=(1e-24, 1e-20),
    norm="log",
    label=r"Strain noise [$1/\sqrt{\mathrm{Hz}}$]",
)
plot
```

Aqui podemos ver como o ASD para LIGO-Livingston evolui ao longo de um período de aproximadamente 17 minutos em torno de GW170817. Estão envolvidas quatro ordens de magnitude na intensidade do ASD, e é por isso que a tentativa anterior resultou em uma imagem azul (quase) totalmente escura.

Podemos ver que o ruído de baixa frequência (<30 Hz) ressoa com alguma variação, mas as altas frequências (>100 Hz) são relativamente estáveis. Entre 30-100 Hz podemos ver algumas características estreitas aparecendo e desaparecendo à medida que o ruído não estacionário afeta a medição.
